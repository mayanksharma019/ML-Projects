```python
#import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#for equations to plot
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#using figure instance
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.2,.2])
axes1.plot(x,y)
axes2.plot(x,y)

```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.4,.4])
#using limits
axes1.set_xlim(0,100)
axes2.set_xlim(20,22)
axes2.set_ylim(30,50)
axes1.plot(x,z)
axes2.plot(x,y)

```

```python
#subplots
plt.subplots(1,2)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,z)
plt.tight_layout()
```

```python
plt.subplots(1,2,figsize=(15,2))
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(x,z)
plt.tight_layout()
```
